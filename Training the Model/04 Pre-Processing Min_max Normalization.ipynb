```python
TRAIN = TRAIN.sample(frac=1);
VALIDATE = VALIDATE.sample(frac=1);
TEST = TEST.sample(frac=1);
```

```python
TRAIN
```

```python
train_labels_1 = TRAIN.pop('Cell_Line')
train_labels_2 = TRAIN.pop('Cancer_Type')
test_labels_1 = TEST.pop('Cell_Line')
test_labels_2 = TEST.pop('Cancer_Type');
validate_labels_1 = VALIDATE.pop('Cell_Line');
validate_labels_2 = VALIDATE.pop('Cancer_Type');
```

```python
columns = np.array(VALIDATE.columns)
```

```python
np.save('columns.npy',columns)
```

```python
lab1_dict = {};
i = 0 ;
for each in train_labels_1.unique():
    lab1_dict[each] = i
    i += 1
    
```

```python
lab2_dict = {};
i = 0 ;
for each in train_labels_2.unique():
    lab2_dict[each] = i
    i += 1
```

```python
lab2_dict
```

```python
lab1_arr = []
for key,values in lab1_dict.items():
    lab1_arr.append([key,values])
```

```python
lab1_arr = np.array(lab1_arr)
np.save('lab1_arr', lab1_arr)
```

```python
lab2_arr = []
for key,values in lab2_dict.items():
    lab2_arr.append([key,values])
    
lab2_arr = np.array(lab2_arr);
np.save('lab2_arr', lab2_arr);
```

```python
lab1_arr = np.load('lab1_arr.npy')
lab2_arr = np.load('lab2_arr.npy')
```

```python
for key,value in lab1_arr:
    train_labels_1 = train_labels_1.replace({key:value})
    test_labels_1 = test_labels_1.replace({key:value})
    validate_labels_1 = validate_labels_1.replace({key:value})
```

```python
for key,value in lab2_arr:
    train_labels_2 = train_labels_2.replace({key:value})
    test_labels_2 = test_labels_2.replace({key:value})
    validate_labels_2 = validate_labels_2.replace({key:value})
```

```python
mean_std_col = np.load('mean_std_cols.npy',allow_pickle=True)
```

```python
df_train_z = TRAIN.copy()
df_test_z = TEST.copy()
df_val_z = VALIDATE.copy()
for key,value in mean_std_col:
    Mean,Std = value[0],value[1]
    df_train_z[key] = (df_train_z[key] - Mean) / Std 
    df_test_z[key] = (df_test_z[key] - Mean) / Std ;
    df_val_z[key] = (df_val_z[key] - Mean) / Std 
    
     
```

```python
min_max_col = np.load('min_max_cols.npy',allow_pickle=True)
```

```python
df_train_s = TRAIN.copy()
df_test_s = TEST.copy()
df_val_s = VALIDATE.copy()
for key,value in min_max_col:
    Min,Max = value[0],value[1]
    df_train_s[key] = (df_train_s[key] - Min) / (Max - Min)  
    df_test_s[key] = (df_test_s[key] - Min) / (Max - Min) ;
    df_val_s[key] = (df_val_s[key] - Min) / (Max - Min) 
```
