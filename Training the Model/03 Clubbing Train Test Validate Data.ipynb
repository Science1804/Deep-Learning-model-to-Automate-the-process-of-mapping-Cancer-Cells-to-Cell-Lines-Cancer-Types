```python
for file in files:  # now repeating the same process we did for one file to all the files in the folder seperated  .. so for each cell line will have now 3 files .
    f = os.path.join('Cell_lines',file)
    rf = pd.read_csv(f);
    rf.rename(columns={'Unnamed: 0':'GENE'},inplace=True);
    rf = rf.set_index('GENE')
    rf = rf.T
    cell_line = file.rstrip('.csv')
    rf.insert(0,'Cell_Line',[cell_line for i in range(len(rf))])
    c_type = labels.loc[cell_line]['cancer_type']
    rf.insert(1,'Cancer_Type',[c_type for i in range(len(rf))])
    
    train, test_val = train_test_split(rf, test_size=0.4, random_state=18, shuffle=True)
    train_path = os.path.join('TTV/Train',file)
    train.to_csv(train_path,sep=',', index=False)
    
    test, validate = train_test_split(test_val, test_size=0.5, random_state=18, shuffle=True)
    test_path = os.path.join('TTV/Test',file);
    validate_path = os.path.join('TTV/Validate',file)
    test.to_csv(test_path,sep=',', index=False);
    validate.to_csv(validate_path,sep=',', index=False);
    
```
